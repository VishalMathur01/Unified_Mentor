# Graphical EDA

```python
sns.histplot(x=data['Units Sold'])
```

```python
data.plot(kind='scatter', x='Units Sold', y='Unit Price', s=32, alpha=.8);
```

```python
sns.histplot(x = data['Unit Price'])
```

```python
sns.histplot(x = data['Unit Cost'])
```

```python
data.plot(kind='scatter', x='Unit Cost', y='Total Revenue', s=32, alpha=.8)
```

```python
data.groupby('Sales Channel').size().plot(kind='barh')
```

```python
data.groupby('Order Priority').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
```

```python
plt.subplots(figsize=(8, 8))
plotter = pd.DataFrame({
    x_label: grp['Sales Channel'].value_counts()
    for x_label, grp in data.groupby('Region')
})
sns.heatmap(plotter, cmap='viridis')
plt.xlabel('Region')
_ = plt.ylabel('Sales Channel')
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Order Priority'].value_counts()
    for x_label, grp in data.groupby('Sales Channel')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Sales Channel')
_ = plt.ylabel('Order Priority')
```
